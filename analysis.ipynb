```python
import os
import numpy as np
import pandas as pd
```

```python
excel_data_filename = r'Basis State Industry List 071621.xlsx'
excel_industries_filename = r'Basis State Industries 210701.xlsx'
excel_industries_range = r'B6:B105'
```

```python
training_df = pd.read_excel(os.path.join('data', excel_data_filename), 
                            sheet_name=r'Training List',
                            engine='openpyxl')
training_df
```

```python
all_data_df = pd.read_excel(os.path.join('data', excel_data_filename), 
                            sheet_name=r'All Acquirers',
                            engine='openpyxl')
all_data_df
```

```python
from openpyxl import load_workbook

industries = []
wb = load_workbook(os.path.join('data', excel_industries_filename))
ws = wb[r'Industries']
for row in ws[excel_industries_range]:
    for cell in row:
        industries.append(cell.value)
        
industries
```
