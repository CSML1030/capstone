## Data modelling of training data

Use NLP techniques for feature engineering

```python
from ftfy import fix_text
import nltk
from nltk.stem.snowball import SnowballStemmer
from nltk.tokenize.casual import casual_tokenize
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS as sklearn_stop_words
```

```python
import re

stemmer = SnowballStemmer("english")
nltk.download("wordnet")
from nltk.stem import WordNetLemmatizer

from collections import defaultdict
lemmatizer = WordNetLemmatizer()

def decontracted(phrase):
    # Taken from https://stackoverflow.com/questions/19790188/expanding-english-language-contractions-in-python

    phrase = re.sub(r"won\'t", "will not", phrase)
    phrase = re.sub(r"can\'t", "can not", phrase)
    phrase = re.sub(r"n\'t", " not", phrase)
    phrase = re.sub(r"\'re", " are", phrase)
    phrase = re.sub(r"\'s", " is", phrase)
    phrase = re.sub(r"\'d", " would", phrase)
    phrase = re.sub(r"\'ll", " will", phrase)
    phrase = re.sub(r"\'t", " not", phrase)
    phrase = re.sub(r"\'ve", " have", phrase)
    phrase = re.sub(r"\'m", " am", phrase)
    return phrase

def tokenize_phrase(text):
    text = str(text)
    text = fix_text(text)
    text = text.encode("ascii", errors="ignore").decode() #remove non ascii chars
    phrase = decontracted(text)
    tokens = casual_tokenize(phrase, reduce_len=True, strip_handles=True)
    normalized_tokens = [x.lower() for x in tokens]  #  convert to all lowercase
    filtered_tokens = [x for x in normalized_tokens if x not in sklearn_stop_words]  #  filter stop words
    filtered_tokens = [x for x in filtered_tokens if x and x not in "- \t\n.\"':[...][\\]()/[br]<>*~,;!?"]  #  filter punctuations
    lemmatized_tokes = [lemmatizer.lemmatize(t) for t in tokens]  # perform lemmatization
    stemmed_tokens = [stemmer.stem(w) for w in filtered_tokens]  # perform stemming
    pos_tokens = nltk.pos_tag(stemmed_tokens)  # perform part-of-speech tagging
    return pos_tokens
```

```python
training_df['description_tokens'] = training_df['Organization - CB Description'].apply(tokenize_phrase)
training_df['industry_tokens'] = training_df['Organization - CB Industries'].apply(tokenize_phrase)
training_df['keyword_tokens'] = training_df['Keyword'].apply(tokenize_phrase)
training_df['metakeyword_tokens'] = training_df['MetaKeyword'].apply(tokenize_phrase)
training_df['metadescription_tokens'] = training_df['MetaDescription'].apply(tokenize_phrase)
training_df['web_description_tokens'] = training_df['web_description'].apply(tokenize_phrase)
training_df['web_keywords_tokens'] = training_df['web_keywords'].apply(tokenize_phrase)
training_df
```

## Construct Bag of Words from tokens

```python
# merge all tokens into 1 column of the DataFrame
tokens_sr = []
for desc_tokens, ind_tokens, key_tokens, metakey_tokens, metadesc_tokens, web_desc_tokens, web_key_tokens in zip(training_df['description_tokens'], training_df['industry_tokens'], training_df['keyword_tokens'], training_df['metakeyword_tokens'], training_df['metadescription_tokens'], training_df['web_description_tokens'], training_df['web_keywords_tokens']):
    tokens = []
    tokens.extend(desc_tokens) if desc_tokens != [('nan', 'NN')] else False
    tokens.extend(ind_tokens) if ind_tokens != [('nan', 'NN')] else False
    tokens.extend(key_tokens) if key_tokens != [('nan', 'NN')] else False
    tokens.extend(metakey_tokens) if metakey_tokens != [('nan', 'NN')] else False
    tokens.extend(metadesc_tokens) if metadesc_tokens != [('nan', 'NN')] else False
    tokens.extend(web_desc_tokens) if web_desc_tokens != [('nan', 'NN')] else False
    tokens.extend(web_key_tokens) if web_key_tokens != [('nan', 'NN')] else False
    tokens_sr.append(tokens)
    
training_df['tokens'] = tokens_sr
training_df
```

```python
# construct bag of words from tokens
from collections import Counter

bag_of_words = []
for tokens in training_df['tokens']:
    bag_of_words.append(Counter(tokens))
    
bows_df = pd.DataFrame.from_records(bag_of_words)
bows_df = bows_df.fillna(0).astype(int)
bows_df
```

```python
# construct DataFrame of all features from bag of words and 1 hot encoding of size
feature_df = pd.concat([bows_df, training_df[['size_1-10', 'size_11-50', 'size_51-100', 'size_101-250', 'size_251-500', 'size_501-1000', 'size_1001-5000', 'size_5001-10000', 'size_10001+']]], axis=1)
feature_df
```

## Construct TF-IDF features set

```python
def sanitize_phrase(text):
    text = str(text)
    text = fix_text(text)
    text = text.encode("ascii", errors="ignore").decode() #remove non ascii chars
    phrase = decontracted(text)
    tokens = casual_tokenize(phrase, reduce_len=True, strip_handles=True)
    normalized_tokens = [x.lower() for x in tokens]  #  convert to all lowercase
    filtered_tokens = [x for x in normalized_tokens if x not in sklearn_stop_words]  #  filter stop words
    filtered_tokens = [x for x in filtered_tokens if x and x not in "- \t\n.\"':[...][\\]()/[br]<>*~,;!?"]  #  filter punctuations
    lemmatized_tokes = [lemmatizer.lemmatize(t) for t in tokens]  # perform lemmatization
    stemmed_tokens = [stemmer.stem(w) for w in filtered_tokens]  # perform stemming
    return " ".join(stemmed_tokens)
```

```python
# tf-idf does not play well with part-of-speech tagging so we need plain text
training_df['sani_desc'] = training_df['Organization - CB Description'].apply(sanitize_phrase)
training_df['sani_indust'] = training_df['Organization - CB Industries'].apply(sanitize_phrase)
training_df['sani_key'] = training_df['Keyword'].apply(sanitize_phrase)
training_df['sani_metakey'] = training_df['MetaKeyword'].apply(sanitize_phrase)
training_df['sani_metadesc'] = training_df['MetaDescription'].apply(sanitize_phrase)
training_df['sani_webdesc'] = training_df['web_description'].apply(sanitize_phrase)
training_df['sani_webkey'] = training_df['web_keywords'].apply(sanitize_phrase)
training_df
```

```python
training_df['sanitized'] = training_df['sani_desc'].apply(lambda x: '' if x == 'nan' else x)
training_df['sanitized'] = training_df['sanitized'] + training_df['sani_indust'].apply(lambda x: '' if x == 'nan' else ' ' + x)
training_df['sanitized'] = training_df['sanitized'] + training_df['sani_key'].apply(lambda x: '' if x == 'nan' else ' ' + x)
training_df['sanitized'] = training_df['sanitized'] + training_df['sani_metakey'].apply(lambda x: '' if x == 'nan' else ' ' + x)
training_df['sanitized'] = training_df['sanitized'] + training_df['sani_metadesc'].apply(lambda x: '' if x == 'nan' else ' ' + x)
training_df['sanitized'] = training_df['sanitized'] + training_df['sani_webdesc'].apply(lambda x: '' if x == 'nan' else ' ' + x)
training_df['sanitized'] = training_df['sanitized'] + training_df['sani_webkey'].apply(lambda x: '' if x == 'nan' else ' ' + x)
training_df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tv = TfidfVectorizer(min_df=0., max_df=1., use_idf=True)
tv_matrix = tv.fit_transform(training_df['sanitized'])
tv_matrix = tv_matrix.toarray()
vocab = tv.get_feature_names()
tfidf_df = pd.DataFrame(np.round(tv_matrix, 2), columns=vocab)
tfidf_df
```

```python
# construct DataFrame of all features from TF-IDF and 1 hot encoding of size
tfidf_feature_df = pd.concat([tfidf_df, training_df[['size_1-10', 'size_11-50', 'size_51-100', 'size_101-250', 'size_251-500', 'size_501-1000', 'size_1001-5000', 'size_5001-10000', 'size_10001+']]], axis=1)
tfidf_feature_df
```

## Stacking with prior models and Bag of Words

```python
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import KFold
from sklearn.naive_bayes import MultinomialNB
from sklearn.neural_network import MLPClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC

dependent_df = training_df['Organization - New Industry']
independent_df = feature_df
x_train, x_test, y_train, y_test = train_test_split(independent_df, dependent_df, test_size=0.2, random_state=5)

x_train = x_train.to_numpy()
x_test = x_test.to_numpy() 
y_train = y_train.to_numpy()
y_test = y_test.to_numpy()

base_learners = []

nb = MultinomialNB()
base_learners.append(nb)

adaboost = AdaBoostClassifier(n_estimators=400, learning_rate=1, algorithm='SAMME')
base_learners.append(adaboost)

rf = RandomForestClassifier(n_estimators=1000, 
                            criterion="gini", 
                            max_depth=100, 
                            min_samples_split=3, 
                            min_samples_leaf=2,
                            random_state=42)
base_learners.append(rf)

svc = SVC(random_state=42, probability=True)
base_learners.append(svc)

bagging = BaggingClassifier(base_estimator=DecisionTreeClassifier(),
                             n_estimators=ensemble_size,
                             oob_score=True)
base_learners.append(bagging)

mlp = MLPClassifier(hidden_layer_sizes =(100, ), solver='lbfgs', random_state=42)
base_learners.append(mlp)

meta_learner = LogisticRegression(solver='lbfgs')
meta_data = np.zeros((len(base_learners), len(x_train)))
meta_targets = pd.Series(np.zeros(len(x_train)))

# Create the cross-validation folds
KF = KFold(n_splits=5)
meta_index = 0
for train_indices, test_indices in KF.split(x_train):
    # Train each learner on the K-1 folds and create meta data for the Kth fold
    for i in range(len(base_learners)):
        learner = base_learners[i]

        learner.fit(x_train[train_indices], y_train[train_indices])
        predictions = learner.predict_proba(x_train[test_indices])[:,0]

        meta_data[i][meta_index:meta_index+len(test_indices)] = predictions

    meta_targets[meta_index:meta_index+len(test_indices)] = y_train[test_indices]
    meta_index += len(test_indices)
    
# Transpose the meta data to be fed into the meta learner
meta_data = meta_data.transpose()

# Create the meta data for the test set and evaluate the base learners
test_meta_data = np.zeros((len(base_learners), len(x_test)))
base_acc = []
for i in range(len(base_learners)):
    learner = base_learners[i]
    learner.fit(x_train, y_train)
    predictions = learner.predict_proba(x_test)[:,0]
    test_meta_data[i] = predictions

    acc = metrics.accuracy_score(y_test, learner.predict(x_test))

    base_acc.append(acc)
    
test_meta_data = test_meta_data.transpose()

# Fit the meta learner on the train set and evaluate it on the test set
meta_learner.fit(meta_data, meta_targets)
ensemble_predictions = meta_learner.predict(test_meta_data)

acc = metrics.accuracy_score(y_test, ensemble_predictions)

# Print the results
print('Acc  Name')
print('-'*20)
for i in range(len(base_learners)):
    learner = base_learners[i]

    print(f'{base_acc[i]:.2f} {learner.__class__.__name__}')
print(f'{acc:.2f} Ensemble')
```
